```python
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=74, value='Black')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=75, value='American Indian/Alaska Native')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=76, value='Asian or Pacific Islander')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=77, value='Filipino')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=78, value='Hispanic')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=79, value='Hawaiian Native/Pacific Islander')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=80, value='White')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=144, value='Two or More Races')
```

```python
#Melt the data to have each ethnicity 
DFfinMelt = pd.melt(df, id_vars=['District Code', 'Agency Name', 'Total Enrollment', 'Total Revenue', 'Revenue per student'], 
        value_vars=['Black', 'Hispanic', 'White', 'Asian or Pacific Islander',
       'American Indian/Alaska Native', 'Hawaiian Native/Pacific Islander','Two or More Races'])
```

```python
#Rename col names to match the DFass
DFfinMelt.columns = ['District Code', 'Agency Name', 'Total Enrollment', 'Total Revenue','Revenue per student', 'Subgroup ID', 'Count Enrollment per ethnicity']
```

```python
DFfinMelt.head(10)
```

```python
#DFfinMelt.to_csv('fin_melt.csv')
```

```python
final_ethnicity_merged = pd.merge(DFfinMelt, DFass, how='outer', on=['District Code', 'Subgroup ID'])
final_ethnicity_merged= final_ethnicity_merged.sort_values(by=['District Code'])
```

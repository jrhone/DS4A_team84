```python
#Convert the amounts to percentages now.
df_final_nonLA['Percent Standard Exceeded'] =(df_final_nonLA['Total Standard Exceeded']/df_final_nonLA['Students Tested']*100).round(decimals=0)
df_final_nonLA['Percent Standard Met'] =(df_final_nonLA['Total Standard Met']/df_final_nonLA['Students Tested']*100).round(decimals=0)
df_final_nonLA['Percent Standard Met and Above'] =(df_final_nonLA['Total Standard Met and Above']/df_final_nonLA['Students Tested']*100).round(decimals=0)
df_final_nonLA['Percent Standard Nearly Met'] =(df_final_nonLA['Total Standard Nearly Met']/df_final_nonLA['Students Tested']*100).round(decimals=0)
df_final_nonLA['Percent Standard Not Met'] =(df_final_nonLA['Total Standard Not Met']/df_final_nonLA['Students Tested']*100).round(decimals=0)
df_final_nonLA.head(40)
```

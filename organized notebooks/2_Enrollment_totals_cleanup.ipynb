# Fetching the relevant data
## Data wrangling

The process of transforming and mapping data from one "raw" data form into another format with the intent of making it more appropriate and valuable for a variety of downstream purposes such as analytics.

The goal here is to join the 3 datasets containing enrollment from prek to grade 12 into one dataframe.

# Prekindergarten to grade 5:

```python
# load the data into a dataframe

df_pk_g5 = pd.read_csv('ELSI_California_district_enrollment_ethnicity_2016_2017-PK_G5.csv')
df_pk_g5.head(10)
```

```python
df_pk_g5.columns
```

```python
df_pk5_race = df_pk_g5.iloc[:,2:].apply(pd.to_numeric, errors='coerce')
df_pk5_race
```

```python
df_pk5_race['Total Prek-5 American Indian/Alaska Native'] = df_pk5_race[['Prekindergarten Students - American Indian/Alaska Native [District] 2016-17',
                                                                     'Kindergarten Students - American Indian/Alaska Native [District] 2016-17',
                                                                     'Grade 1 Students - American Indian/Alaska Native [District] 2016-17',
                                                                     'Grade 2 Students - American Indian/Alaska Native [District] 2016-17',
                                                                     'Grade 3 Students - American Indian/Alaska Native [District] 2016-17',
                                                                     'Grade 4 Students - American Indian/Alaska Native [District] 2016-17',
                                                                     'Grade 5 Students - American Indian/Alaska Native [District] 2016-17']].sum(axis=1)
df_pk5_race.head()
```

```python
df_pk5_race['Total Prek-5 Asian or Asian/Pacific Islander'] = df_pk5_race[['Prekindergarten Students - Asian or Asian/Pacific Islander [District] 2016-17',
                                                                     'Kindergarten Students - Asian or Asian/Pacific Islander [District] 2016-17',
                                                                     'Grade 1 Students - Asian or Asian/Pacific Islander [District] 2016-17',
                                                                     'Grade 2 Students - Asian or Asian/Pacific Islander [District] 2016-17',
                                                                     'Grade 3 Students - Asian or Asian/Pacific Islander [District] 2016-17',
                                                                     'Grade 4 Students - Asian or Asian/Pacific Islander [District] 2016-17',
                                                                     'Grade 5 Students - Asian or Asian/Pacific Islander [District] 2016-17']].sum(axis=1)
df_pk5_race.head()
```

```python
df_pk5_race['Total Prek-5 Black'] = df_pk5_race[['Prekindergarten Students - Black [District] 2016-17',
                                                                     'Kindergarten Students - Black [District] 2016-17',
                                                                     'Grade 1 Students - Black [District] 2016-17',
                                                                     'Grade 2 Students - Black [District] 2016-17',
                                                                     'Grade 3 Students - Black [District] 2016-17',
                                                                     'Grade 4 Students - Black [District] 2016-17',
                                                                     'Grade 5 Students - Black [District] 2016-17']].sum(axis=1)
df_pk5_race.head()
```

```python
df_pk5_race['Total Prek-5 Hispanic'] = df_pk5_race[['Prekindergarten Students - Hispanic [District] 2016-17',
                                                                     'Kindergarten Students - Hispanic [District] 2016-17',
                                                                     'Grade 1 Students - Hispanic [District] 2016-17',
                                                                     'Grade 2 Students - Hispanic [District] 2016-17',
                                                                     'Grade 3 Students - Hispanic [District] 2016-17',
                                                                     'Grade 4 Students - Hispanic [District] 2016-17',
                                                                     'Grade 5 Students - Hispanic [District] 2016-17']].sum(axis=1)
df_pk5_race.head()
```

```python
df_pk5_race['Total Prek-5 White'] = df_pk5_race[['Prekindergarten Students - White [District] 2016-17',
                                                                     'Kindergarten Students White [District] 2016-17',
                                                                     'Grade 1 Students - White [District] 2016-17',
                                                                     'Grade 2 Students - White [District] 2016-17',
                                                                     'Grade 3 Students - White [District] 2016-17',
                                                                     'Grade 4 Students - White [District] 2016-17',
                                                                     'Grade 5 Students - White [District] 2016-17']].sum(axis=1)
df_pk5_race.head()
```

```python
df_pk5_race['Total Prek-5 Hawaiian Nat./Pacific Isl.'] = df_pk5_race[['Prekindergarten Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Kindergarten Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Grade 1 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Grade 2 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Grade 3 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Grade 4 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Grade 5 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17']].sum(axis=1)
df_pk5_race.head()
```

```python
df_pk5_race['Total Prek-5 Two or More Races'] = df_pk5_race[['Prekindergarten Students Two or More Races [District] 2016-17',
                                                                     'Kindergarten Students Two or More Races [District] 2016-17',
                                                                     'Grade 1 Students - Two or More Races [District] 2016-17',
                                                                     'Grade 2 Students - Two or More Races [District] 2016-17',
                                                                     'Grade 3 Students - Two or More Races [District] 2016-17',
                                                                     'Grade 4 Students - Two or More Races [District] 2016-17',
                                                                     'Grade 5 Students - Two or More Races [District] 2016-17']].sum(axis=1)
df_pk5_race.head()
```

```python
df_pk5_totals = df_pk5_race.drop(df_pk5_race.columns[0:-7], axis = 1)
df_pk5_totals
```

```python
df_pk5_totals['Total Enrollment pk-5'] = df_pk5_totals.sum(axis=1)
df_pk5_totals
```

```python
agency_name = df_pk_g5.iloc[:, :1]
agency_name
```

```python
total_enrollment_pk5 = pd.merge(agency_name, df_pk5_totals, left_index=True, right_index=True)
total_enrollment_pk5
```

## Below a table contaning total enrollment by race from prek to grade 9:

```python
enrollment_pk_9 = pd.merge(total_enrollment_pk5, df_g6to9_totals, left_index=True, right_index=True)
enrollment_pk_9
```

```python
enrollment_pk_9.columns
```

```python
enrollment_pk_9['Total American Indian/Alaska Native'] = enrollment_pk_9[['Total Prek-5 American Indian/Alaska Native', 
                                                                   'Total grade 6-9 American Indian/Alaska Native']].sum(axis=1)
enrollment_pk_9['Total Black'] = enrollment_pk_9[['Total Prek-5 Black','Total grade 6-9 Black']].sum(axis=1)
enrollment_pk_9['Total Hispanic'] = enrollment_pk_9[['Total Prek-5 Hispanic','Total grade 6-9 Hispanic']].sum(axis=1)
enrollment_pk_9['Total White'] = enrollment_pk_9[['Total Prek-5 White','Total grade 6-9 White']].sum(axis=1)
enrollment_pk_9['Total Hawaiian Nat./Pacific Isl.'] = enrollment_pk_9[['Total Prek-5 Hawaiian Nat./Pacific Isl.',
                                                                 'Total grade 6-9 Hawaiian Nat./Pacific Isl.']].sum(axis=1)
enrollment_pk_9['Total Two or More Races'] = enrollment_pk_9[['Total Prek-5 Two or More Races','Total grade 6-9 Two or More Races']].sum(axis=1)
enrollment_pk_9['Total Asian or Asian/Pacific Islander'] = enrollment_pk_9[['Total Prek-5 Asian or Asian/Pacific Islander','Total grade 6-9 Asian or Asian/Pacific Islander']].sum(axis=1)

enrollment_pk_9.head()
```

```python
total_enrollment_byrace = enrollment_pk_9.drop(enrollment_pk_9.columns[1:-7], axis = 1)
total_enrollment_byrace['Total Enrollment'] = total_enrollment_byrace.iloc[:, 1:].sum(axis=1)
total_enrollment_byrace
```

# Grade 10 to grade 12:


```python
# load the data into a dataframe

df_g10_g12 = pd.read_csv('ELSI_California_district_enrollment_ethnicity_2016_2017_G10-G12.csv')
df_g10_g12.head(10)
```

```python
df_total['Total G10-G12 Black'] = df_total[['Grade 10 Students - Black [District] 2016-17',
                                                                     'Grade 11 Students - Black [District] 2016-17',
                                                                     'Grade 12 Students - Black [District] 2016-17']].sum(axis=1)
df_total.head() 
```

```python
df_total['Total G10-G12 White'] = df_total[['Grade 10 Students - White [District] 2016-17',
                                                                     'Grade 11 Students - White [District] 2016-17',
                                                                     'Grade 12 Students - White [District] 2016-17']].sum(axis=1)
df_total.head() 
```

```python
df_total['Total G10-G12 Hawaiian Nat./Pacific Isl.'] = df_total[['Grade 10 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Grade 11 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17',
                                                                     'Grade 12 Students - Hawaiian Nat./Pacific Isl. [District] 2016-17']].sum(axis=1)
df_total.head() 
```

```python
df_total['Total G10-G12 Two or More Races'] = df_total[['Grade 10 Students - Two or More Races [District] 2016-17',
                                                                     'Grade 11 Students - Two or More Races [District] 2016-17',
                                                                     'Grade 12 Students - Two or More Races [District] 2016-17']].sum(axis=1)
df_total.head() 
```

```python
total_enrollment_byrace
```

```python
total_enrollment = pd.merge(total_enrollment_byrace, df_total, left_index=True, right_index=True)
total_enrollment
```

```python
total_enrollment['American Indian/Alaska Native'] = total_enrollment[['Total American Indian/Alaska Native', 'Total G10-G12 American Indian/Alaska Native']].sum(axis=1)
total_enrollment['Black'] = total_enrollment[['Total Black', 'Total G10-G12 Black']].sum(axis=1)
total_enrollment['Hispanic'] = total_enrollment[['Total Hispanic', 'Total G10-G12 Hispanic']].sum(axis=1)
total_enrollment['White'] = total_enrollment[['Total White', 'Total G10-G12 White']].sum(axis=1)
total_enrollment['Hawaiian Nat./Pacific Isl.'] = total_enrollment[['Total Hawaiian Nat./Pacific Isl.', 'Total G10-G12 Hawaiian Nat./Pacific Isl.']].sum(axis=1)
total_enrollment['Two or More Races'] = total_enrollment[['Total Two or More Races', 'Total G10-G12 Two or More Races']].sum(axis=1)
total_enrollment['Asian or Asian/Pacific Islander'] = total_enrollment[['Total Asian or Asian/Pacific Islander', 'Total G10-G12 Asian or Asian/Pacific Islander']].sum(axis=1)


total_enrollment

```

```python
total_enrollment_new = total_enrollment.drop(total_enrollment.columns[1:-7], axis = 1)
total_enrollment_new
```

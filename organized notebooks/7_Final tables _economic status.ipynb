```python
#Disadvantaged label
def define_label(x):
    if x in (200,201,202,203,204,205,206,207):
        return 'Disadvantaged'
    else:
        return 'Not Disadvantaged'
DFass['Disadvantaged']=DFass['Subgroup ID'].apply(define_label)
DFass
```

```python
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=200, value='Black')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=201, value='American Indian/Alaska Native')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=202, value='Asian or Asian/Pacific Islander')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=203, value='Filipino')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=204, value='Hispanic')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=205, value='Hawaiian Nat./Pacific Isl.')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=206, value='White')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=207, value='Two or More Races')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=220, value='Black')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=221, value='American Indian/Alaska Native')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=222, value='Asian or Asian/Pacific Islander')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=223, value='Filipino')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=224, value='Hispanic')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=225, value='Hawaiian Nat./Pacific Isl.')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=226, value='White')
DFass['Subgroup ID']=DFass['Subgroup ID'].replace(to_replace=227, value='Two or More Races')
```

```python
#Melt the data to have each ethnicity 
DFfinMelt = pd.melt(df, id_vars=['District Code', 'Agency Name', 'Total Enrollment', 'Total Revenue', 'Revenue per student'], 
        value_vars=['Black', 'Hispanic', 'White', 'Asian or Asian/Pacific Islander','American Indian/Alaska Native', 'Hawaiian Nat./Pacific Isl.','Two or More Races'])
```

```python
#Rename col names to match the DFass
DFfinMelt.columns = ['District Code', 'Agency Name', 'Total Enrollment', 'Total Revenue','Revenue per student', 'Subgroup ID', 'Count Enrollment per ethnicity']
```

```python
final_eco_ethnicity_merged = pd.merge(DFfinMelt, DFass, how='outer', on=['District Code', 'Subgroup ID'])
final_eco_ethnicity_merged= final_eco_ethnicity_merged.sort_values(by=['District Code'])
```

```python
final_eco_ethnicity_merged=final_eco_ethnicity_merged.dropna()
```
